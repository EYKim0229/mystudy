```r
geneSample <- read.csv( 'train_data_geneBySample.csv' , header=T)
```

```r
head( geneSample )
summary( geneSample  )
```

- 마지막 71라인의 데이터가 문제가 있음
- 마지막 라인은 읽지 않도록 처리가 필요함.
- read.csv()에서 comment.char = "A" 라는 옵션을 추가해서 마지막 라인은 읽지 않도록함. 위험한 방법임. 나중에는 좀더 우아한 방법을 찾아보자.
- probeName은 rowname 으로 처리함.

```r
geneSampleClean <- read.csv( 'train_data_geneBySample.csv' , header=T, comment.char = "A", row.names="probeName" )
tail( geneSampleClean )
```

```r
dim( geneSampleClean )
summary( geneSampleClean  )
```

```r
boxplot( geneSampleClean )
```

- boxplot()결과 모든 데이터가 평균과 분산이 비슷하므로, 표준화가 필요없음.

## 2. Distance 

```r
geneSampleClean.dist <- dist( geneSampleClean, method="euclidean"  ) 
geneSampleClean.dist
```

```r
geneSampleClean.dist.t <- dist( t(geneSampleClean), method="euclidean"  ) 
geneSampleClean.dist.t
```

```r
geneSampleClean.manh <- dist( geneSampleClean, method="manhattan"  ) 
geneSampleClean.manh
```

```r
geneSampleClean.manh.t <- dist( t(geneSampleClean), method="manhattan"  ) 
geneSampleClean.manh.t
```

```r
geneSampleClean.cor  <- dist( 1 - cor(geneSampleClean) )
geneSampleClean.cor
```

##  3. Clustering analysis and plot

```r
plot( hclust(geneSampleClean.dist) )
```

```r
#par( mfrow=c(1, 2) )
plot( hclust(geneSampleClean.dist, method="single") )
plot( hclust(geneSampleClean.dist.t, method="single") )
plot( hclust(geneSampleClean.dist.t, method="average") )
plot( hclust(geneSampleClean.dist.t, method="complete") )
plot( hclust(geneSampleClean.dist.cor, method="average") )
#par( mfrow=c(1, 1) )
```

##  4. Heatmap analysis 

- heatmap()함수가 matrix형만 지원하므로, data.frame형을 matrix형으로 변환.

```r
class(geneSampleClean )
```

```r
geneSampleMatrix <- as.matrix( geneSampleClean )
class(geneSampleMatrix)
```

```r
heatmap( geneSampleMatrix )
```

```r
heatmap( geneSampleMatrix, col=c("green", "red") )
```

```r
heatmap( geneSampleMatrix, col=c("green", "black" , "red" ) )
```
