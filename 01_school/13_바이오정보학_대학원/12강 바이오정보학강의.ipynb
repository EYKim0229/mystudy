###  12강 온라인 강의에 나온 Phylogenetic Tree를 그리는 과제를 수행하시오.


|     | P53 |  A  |  B  |  C  |  D  |
|-----|-----|-----|-----|-----|-----|
| P53 |  0  |     |     |     |     | 
|  A  |  3  |  0  |     |     |     |
|  B  |  6  |  8  |  0  |     |     |
|  C  |  12 |  6  |  11 |  0  |     |
|  D  |  5  |  7  |  2  |  8  |  0  |


P53과 B 유전자가 거리가 가장 가까우므로, E(P53,B)로 군집을 설정함.

- A <-> E(P53,B) = 1/2 ( 3 + 8 ) = 5.5
- C <-> E(P53,B) = 1/2 ( 12 + 11 ) = 11.5
- D <-> E(P53,B) = 1/2 ( 5 + 2 ) = 3.5

|          | E(P53,B) |  A  |  C  |  D  |
|----------|----------|-----|-----|-----|
| E(P53,B) |    0     |     |     |     | 
|     A    |    5.5   |  0  |     |     |
|     C    |    11.5  |  6  |  0  |     |
|     D    |    3.5   |  7  |  8  |  0  |

E(P53,B)와 A 유전자가 가장 가까우므로, F(P53,A,B)로 군집을 설정함

- C <-> F(P53,A,B) = 1/3 ( 12 + 6 + 11 ) = 9.66
- D <-> F(P53,A,B) = 1/3 ( 5 + 7 + 2 ) = 4.66

|           | E(P53,A,B) |  C  |  D  |
|-----------|------------|-----|-----|
| E(P53,A,B)|     0      |     |     |
|     C     |     9.66   |  0  |     |
|     D     |     4.66   |  8  |  0  |

F(P53,A,B)와 D 유전자가 가장 가까우므로, F(P53,A,B,D)로 군집을 설정함.

- C <-> F(P53,A,B,D) = 1/4 ( 12 + 6 + 11 + 8 ) = 9.25

|           | E(P53,A,B) |  C  |
|-----------|------------|-----|
| E(P53,A,B)|     0      |     |
|     C     |     9.25   |  0  |

### 유전자들이 거리가 문제와 비슷하게 나오도록 설정하고 R의 clust()함수를 이용해서 그리면, 아래와 같이 나옴.

```r
data <- matrix( c(
   0, 0, 0,
   3, 0, 0,
   6, 0, 5,
   12,0, 0,
   5, 1, 0
   ), 
  nrow=5, byrow=TRUE )

data
```

```r
rownames(data) = c("P53", "A", "B", "C", "D")
data
```

```r
d <- dist(data, method="manhattan")
d
hc <- hclust(d) 
plot(hc)
```
