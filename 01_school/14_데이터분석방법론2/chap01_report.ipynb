####   2. 참고문헌에서 제시한 “오즈비와 상대위험도 관련 자료” 중 하나의 논문을 읽고 그 내용과 자기 생각을 A4 용지 1매로 작성하시오

```
INFORMATION POINT : Odds ratio

오즈비는 어떤 요인이 존재할때, 특정 질병의 위험을 평가하는데 사용되는 통계학의 한가지 범위이다. 
오즈비는 어떤 요인에 리스크의 노출되었을때와 노출되지 않았을때의 상대적인 척도이다. 

오즈는 확률로 제시되는 방법이지만, 도박에 대해 잘 알지 못한다면 확률을 계산하는 방법에 대한 설명이 필요하다.
어떤 이벤트가 일어날 오즈는 이벤트가 일어난 확률과 일어나지 않을 확률을 나눈 확률값이다.
예를 들어, 3/4미만으로 식사 ,그 이상의 식사에 대한 독립적으로 또는 의존적인 식사에 대한 결과를 보자.
```

####  3. 두 처리 (A, B)와 반응변수 (사망, 생존)간의 오즈비가 2.0이라고 하자. 

(1) “처리 A의 사망 확률은 처리 B에 비하여 2배이다”라는 해석이 왜 틀렸는지 설명하고 옳은 해석을 하시오. 


```
오즈비 => 처리A의 오즈 / 처리B의 오즈 
      => ( 처리A의 사망 / 처리A의 생존 ) / ( 처리B의 사망 / 처리B의 생존 ) 
      => ( 처리A의 사망 / 처리B의 사망 ) * ( 처리B의 생존 / 처리A의 생존 )
      => ( 처리A와 처리B의 상대 위험율 ) * ( 처리B의 생존 / 처리A의 생존 )
이와 같이 상대위험율 * ( 처리B의 생존 / 처리A의 생존 ) 이기 때문에 2배라고 해석할 수 없으며,

두 처리(A, B)는 반응변수( 사망과, 생존)에 강한 연관성이 있음.
```

(2) (1)에서의 해석이 근사적으로 맞는 경우는 어느 경우인가? 

```
처리(A, B)의 사망률이 극히 낮으면, 생존률이 매우 높기 때문에  ( 처리B의 생존 / 처리A의 생존 ) 값이 1의 근사되기 때문에 오즈비를 상대위험률로 해석이 가능함.
```

(3) 처리 A의 사망 오즈가 0.5일 때 처리 A와 처리 B의 사망 확률을 구하시오.

```
오즈 A( = 0.5 ) / 오즈B = 2.0 이므로  오즈B 는 0.25

오즈A = 처리A의 사망율 / 처리A의 생존율 = 처리A의 사망율 / ( 1- 처리A의 사망율 )
0.5 =  A / ( 1- A)  =>  0.5 -0.5A =  A   =>   1.5A = 0.5  =>  A = 1/3

이와 같이 처리B의 확률은 
0.25 = B / ( 1 - B ) =>  1.25B = 0.25  => B = 1/5 

즉,  처리A의 사망확률은 1/3 이고, 처리B의 사망확률은 1/5임
```
