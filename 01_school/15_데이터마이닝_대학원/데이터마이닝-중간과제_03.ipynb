## 6강 연습문제

### 6-1 Fisher의 붓꽃 데이터는 아래 표와 같다. 
### 150개의 관찰치와 4개의 변수를 모두 이용하여 다음과 같이 군집분석을 실시하라.

```r
data(iris)
head(iris)
```

#### (1) 각 변수의 표준화한 값을 구하라.

```r
iris_scale <- scale(iris[,-5])
iris_scale
```

#### (2) K-평균 군집분석 방법을 이용하여 3개 군집에 대한 군집분석을 실시하라.

```r
iris_scale.hclust_average <- hclust( dist(iris_scale), method="average" )
plot( iris_scale.hclust_average )
```

```r
initial <- tapply( iris_scale, 
                 list( rep( cutree( iris_scale.hclust_average, 3), ncol(iris_scale) ), col(iris_scale) ),
                 mean )

iris_scale.kmeans <- kmeans(iris_scale, initial, algorithm="MacQueen" )
iris_scale.kmeans
```

```r
par(mfrow = c(1, 2))
plot( iris_scale[ , c("Sepal.Length", "Sepal.Width")], col=iris_scale.kmeans$cluster )
points(iris_scale.kmeans$centers[,c("Sepal.Length", "Sepal.Width")], col=1:3, pch = 4, cex = 4, lwd = 4)

plot( iris_scale[ , c("Petal.Length", "Petal.Width")], col=iris_scale.kmeans$cluster )
points(iris_scale.kmeans$centers[,c("Petal.Length", "Petal.Width")], col=1:3, pch = 4, cex = 4, lwd = 4)
par(mfrow = c(1, 1))
```

```r
par(mfrow = c(1, 2))
plot( iris_scale[ , c("Petal.Width", "Sepal.Width")], col=iris_scale.kmeans$cluster )
points(iris_scale.kmeans$centers[,c("Petal.Width", "Sepal.Width")], col=1:3, pch = 4, cex = 4, lwd = 4)

plot( iris_scale[ , c("Petal.Length", "Sepal.Length")], col=iris_scale.kmeans$cluster )
points(iris_scale.kmeans$centers[,c("Petal.Length", "Sepal.Length")], col=1:3, pch = 4, cex = 4, lwd = 4)
par(mfrow = c(1, 1))
```

#### (3) K-평균 군집분석의 군집결과와 붓꽃 데이터에 주어져 있는 붓꽃종류와의 분할표를 구성하고 군집분석의 성능에 대해 평가하라.

```r
y = iris$Species
table(y, iris_scale.kmeans$cluster)
```

** 성능평가 **
- setosa는 군집1로 잘 그룹화
- versicolor는 39/50은 군집2으로, 11/50은 군집3으로 잘 그룹화 되지 않음.
- virginica 은 17/50은 군집2으로, 33/50은 군집3으로 잘 그룹화 되지 않음.
- setosa >  versicolor >  virginica 순으로 군집성능이 좋음.

### 6-2 세계의 46개 주요 도시에 대한 물가와 소득에 관한 데이터를 아래와 같이 수집하였다. (1991년 기준) 

### 노동시간, 물가, 소득수준 등의 변수를 사용하여 46개 도시에 대한 군집분석을 다음과 같이 실시하라. 

```r
price_income <- read.csv( '06_02.csv' , head=T )
head( price_income )
```

#### (1) 각 변수의 표준화한 값을 구하라.

```r
rnames <- price_income[ , 1]
rownames(price_income ) <- rnames
price_income <- price_income[ , -1]

price_income.scale <- scale( price_income )
head( price_income.scale )
```

#### (2) 완전연결법을 이용하여 도시들을 군집화하고 수형도로 표현하라. 몇 개의 군집이 적절하다고 판단되는가?

```r
price_income.scale.dist <- dist( price_income.scale , method = "euclidean")
price_income.scale.dist.hclust_complete <- hclust(price_income.scale.dist , method="complete")

plot( price_income.scale.dist.hclust_complete )
rect.hclust(price_income.scale.dist.hclust_complete, k=4, border="red")
```

```r
plot( price_income.scale.dist.hclust_complete )
rect.hclust(price_income.scale.dist.hclust_complete, k=5, border="red")
```

```r
plot( price_income.scale.dist.hclust_complete )
rect.hclust(price_income.scale.dist.hclust_complete, k=6, border="red")
```

```r
plot( price_income.scale.dist.hclust_complete )
rect.hclust(price_income.scale.dist.hclust_complete, k=7, border="red")
```

- 4개로 군집을 나누는 것이 그룹별 속한 수도의 개수가 가장 비슷하여 4개가 적당함

#### (3) Ward 방법을 이용하여 도시들을 군집화하고 수형도로 표현하라. 몇 개의 군집이 적절하다고 판단되는가?

```r
price_income.scale.dist.hclust_ward <- hclust(price_income.scale.dist , method="ward.D2")

plot( price_income.scale.dist.hclust_ward )
rect.hclust(price_income.scale.dist.hclust_ward, k=4, border="red")
```

```r
plot( price_income.scale.dist.hclust_ward )
rect.hclust(price_income.scale.dist.hclust_ward, k=5, border="red")
```

```r
plot( price_income.scale.dist.hclust_ward )
rect.hclust(price_income.scale.dist.hclust_ward, k=6, border="red")
```

- 4개의 그룹으로 나누는 것이 적당해보임.

#### (4) K-평균 군집분석 방법을 이용하여 군집분석을 실시하라. 군집의 수는 (2)번 또는 (3번)의 결과를 이용하여라.

```r
price_income.scale.kmeans <- kmeans(price_income.scale, 4, algorithm="MacQueen" )
price_income.scale.kmeans
```

```r
plot( price_income.scale[ , c("avg_work_time", "prices")], col=price_income.scale.kmeans$cluster )
points(price_income.scale.kmeans$centers[,c("avg_work_time", "prices")], col=1:4, pch = 4, cex = 4, lwd = 4)
```

```r
plot( price_income.scale[ , c("avg_work_time", "income")], col=price_income.scale.kmeans$cluster )
points(price_income.scale.kmeans$centers[,c("avg_work_time", "income")], col=1:4, pch = 4, cex = 4, lwd = 4)
```

```r
plot( price_income.scale[ , c("prices", "income")], col=price_income.scale.kmeans$cluster )
points(price_income.scale.kmeans$centers[,c("prices", "income")], col=1:4, pch = 4, cex = 4, lwd = 4)
```

- ** avg_work_time와 prices 변수가 군집을 구분하는 핵심변수로 파악됨. **
