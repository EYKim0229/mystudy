### 3-2  In the data set vitcap, use a t test to compare the vital capacity for the two groups. Calculate a 99% confidence interval for the difference. The result of this comparison may be misleading. Why?

```r
library( ISwR )
```

```r
summary( vitcap  )
```

```r
# group을 요인으로 변환함.
vitcap2 = transform(vitcap, group=factor(group))
summary( vitcap2  )
```

```r
# group1과 group3을 나누어 추출함.
vitcap2_g1 <- vitcap2[  vitcap2$group == 1, ] 
vitcap2_g3 <- vitcap2[  vitcap2$group == 3, ] 
```

```r
summary( vitcap2_g1 )
summary( vitcap2_g3 )
```

```r
# 상자 그림을 통해서 대략적으로 두 그룹의 분포를 파악하고, group1이 폐활량이 좀 낮은것으로 확인할 수 있음.
boxplot( vital.capacity~group, data=vitcap) 
```

```r
# var.test()d을 통해서 group1과 group3의 폐활량의 분산이 다르다고 할 수 있음
var.test( vitcap2_g1$vital.capacity , vitcap2_g3$vital.capacity  )

# t-test는 분산이 동일하다고 가정해서 실행함.
t.test( vitcap2_g1$vital.capacity ,vitcap2_g3$vital.capacity , val.equal=T , conf.level = 0.99 )
```

- **결과 : t-test 결과 p-value가 0.01보다 작기 때문에 귀무가설인 두 그룹의 평균이 같다라는 가정을 기각하여 두 그룹은 평균이 다름.**

```r
par(mfrow=c(1,2))
hist( vitcap2_g1$vital.capacity,  main ="group1", xlab ="vital.capacity" ) 
hist( vitcap2_g3$vital.capacity,  main ="group3", xlab ="vital.capacity" ) 
par(mfrow=c(1,1))
```

- ** 그러나, 2그룹에 대해서 히스토그램을 그려보면, 두 그룹의 분포가 정규분포라고 할 수 없고, 분포형태도 다르기 때문에 t test 결과를 항상 신뢰할 수 가 없음. ** 
- ** 위의 t-test의 결과는 group1고 group3의 vital.capacity이 다르다는 증거일뿐 group3의 vital.capacity가 크다라는 것은 아님. **

### 3-5  With the rmr data set, plot metabolic rate versus body weight. Fit a linear regression model to the relation. According to the fitted model, what is the predicted metabolic rate for a body weight of 70 kg? Give a 95% confidence interval for the slope of the line.

```r
summary( rmr )
class( rmr )
```

```r
attach( rmr )
rmr.lm <- lm( metabolic.rate ~ body.weight  )

plot( body.weight,  metabolic.rate,  main="metabolic rate versus body weight"  ) 
abline( rmr.lm) 
detach( rmr )
```

```r
summary(rmr.lm)
```

```r
pred_v <- 811.2267 +   7.0595 * 70
pred_v
predict(rmr.lm, newdata=data.frame(body.weight=70))
```

- 몸무게가 70KG 일때 신진대사속도는 약 1305.39로 나옴.

```r
attach( rmr )
new.body_weight <- data.frame( body.weight=40:140 )
pc <- predict(rmr.lm, interval = "confidence", newdata=new.body_weight, level = 0.95 )
pred.weight <- new.body_weight$body.weight

plot( body.weight,  metabolic.rate,  main="metabolic rate versus body weight"  ) 
abline( rmr.lm) 
lines(pred.weight, pc[ ,3], lty = 'dashed', col = 'red')
lines(pred.weight, pc[ ,2], lty = 'dashed', col = 'red')

detach( rmr )
```
