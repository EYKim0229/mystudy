### 3-2  In the data set vitcap, use a t test to compare the vital capacity for the two groups. Calculate a 99% confidence interval for the difference. The result of this comparison may be misleading. Why?

```r
# ISwR 패키지는 한번만 설치하면 됨.
#install.packages(c('ISwR'),  repos="http://cran.nexr.com/"  )
```

```r
library( ISwR )
```

```r
summary( vitcap  )
```

```r
# group을 요인으로 변환함.
vitcap2 = transform(vitcap, group=factor(group))
summary( vitcap2  )
```

```r
# group1과 group3을 나누어 추출함.
vitcap2_g1 <- vitcap2[  vitcap2$group == 1, ] 
vitcap2_g3 <- vitcap2[  vitcap2$group == 3, ] 
```

```r
summary( vitcap2_g1 )
summary( vitcap2_g3 )
```

```r
# 상자 그림을 통해서 대략적으로 두 그룹의 분포를 파악하고, group1이 폐활량이 좀 낮은것으로 확인할 수 있음.
boxplot( vital.capacity~group, data=vitcap) 
```

```r
# var.test()d을 통해서 group1과 group3의 폐활량의 분산이 다르다고 할 수 있음
var.test( vitcap2_g1$vital.capacity , vitcap2_g3$vital.capacity  )

# t-test는 분산이 동일하다고 가정해서 실행함.
t.test( vitcap2_g1$vital.capacity ,vitcap2_g3$vital.capacity , val.equal=T , conf.level = 0.99 )
```

- **결과 : t-test 결과 p-value가 0.01보다 작기 때문에 귀무가설인 두 그룹의 평균이 같다라는 가정을 기각하여 두 그룹은 평균이 다름.**

```r
par(mfrow=c(1,2))
hist( vitcap2_g1$vital.capacity,  main ="group1", xlab ="vital.capacity" ) 
hist( vitcap2_g3$vital.capacity,  main ="group3", xlab ="vital.capacity" ) 
par(mfrow=c(1,1))
```

- ** 그러나, 2그룹에 대해서 히스토그램을 그려보면, 두 그룹-의 분포가 정규분포라고 할 수 없고, 분포형태도 다르기 때문에 t test 결과를 항상 신뢰할 수 가 없음. ** 
- ** 위의 t-test의 결과는 group1고 group3의 vital.capacity이 다르다는 증거일뿐 group1과 group3의 vital.capacity가 크기를 검증하는 것은 아님.  작거나 큰것을 비교하기 위해서는 t-test 의 alternative 옵션을  less 또는 greater을 사용함. **

- index a numeric vector giving an index of mixed epidermal celllymphocyte reactions.
- gvhd a numeric vector code, graft-versus-host disease, 0: no, 1: yes.
- time a numeric vector, follow-up time
- dead a numeric vector code, 0: no (censored), 1: yes


- type , preg , gvhd 는 범주형변수이지만, numberic형으로 되어있어서 범주형으로 변환이 필요함.

```r
ix <- c("preg", "gvhd")
graft.vs.host[ix] <- lapply(graft.vs.host[ix], factor, levels=0:1, labels=c("No","Yes"))
```

```r
graft.vs.host$type <- factor(graft.vs.host$type, labels=c("AML", "ALL", "CML")) 
summary( graft.vs.host )
```

```r
summary(glm(dead ~ rcpage + donage + index + time  , family=binomial, data=graft.vs.host))
```

```r
opar <- par(mfrow=c(2,2))
boxplot( graft.vs.host$index )
boxplot( log( graft.vs.host$index ), main="log transform" )
boxplot( sqrt( graft.vs.host$index ), main="sqrt transform" )
boxplot( ( graft.vs.host$index )^2 , main="^2 transform" )
par( opar )
```

- 여러가지 변환을 해서 log변환시에 정규분포에 가장 가까움.

```r
summary(glm(dead ~ rcpage + donage + log(index) + time  , family=binomial, data=graft.vs.host))
```

- index에 대해서 로그변환 이전과 이후 모두 유의하지 않는것으로 나옴.
- rcpage, donage 변수도 유의하지 않는것으로 나와서 변수를 제거하고 time과 범주형 변수인 type, preg, gvhd에 대해서 모형 적합을 시도함.

```r
summary(glm(dead ~ type + time, family=binomial, data=graft.vs.host))
```

- type이 CML일때만 유의미함.

```r
summary(glm(dead ~ preg + time, family=binomial, data=graft.vs.host))
```

- preg가 No일때만 유의미함.

```r
summary(glm(dead ~ gvhd + time, family=binomial, data=graft.vs.host))
```

- gvhd가 No일때만 유의미함.
