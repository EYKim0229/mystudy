## 9.1 The principles of power calculations

- 잘못된 결론이 나올 오류의 종류
    - 1) The hypothesis is correct, but the test rejects it(type I error). $ \space\space\ \alpha = P( H_1 | H_0 )  $ 
    - 2) The hypothesis is wrong, but the test accepts it(type II error). $ \space\space\ \beta = P ( H_0 | H_1 )  $

- false hypothesis( 귀무가설이 거짓)일때 귀무가설을 기각할 확률을 power( 검정력 )이라 함.
- power $  =  1 - \beta = P( H_1 | H_1 ) $ 

### 9.1.1 Power of one-sample and paired t tests

- paired t-test을 사용해서 처리A와 처리B 사이의 차이가  zero 인지를 확인하고 싶을때,
- 유의미한 차이를 $ \delta $ 라고 하자. 
- null hypothesis( 귀무가설 )이 거짓일때, 검정통계량의 분포는  noncentral t-distribution 임.
- noncentral t distribution는 자유도와 noncentrality parameter 에 의존적임.
- noncentrality parameter  $ v $는  $ \delta $ 와 표준편차 $ \sigma $, 샘플사이즈 $ n $의 함수이다.  
$$ v = \frac{ \delta } {  \sigma / \sqrt{n} }   $$

- v 가 3이고, 자유도가 25인 누적 noncentral t-distribution을 그려보자.
- pt()함수에 ncp 인자로 3을 설정하고, df = 25로 설정하면 됨.

```r
curve(pt(x,25,ncp=3), from=0, to=6)
abline(v=qt(.975,25))
```

```r
pt(qt(.975,25),25,ncp=3)
```

- 채택역( acceptance region )안에 포함된 확률은 0.18 이고, 이것의 검정력은 0.82 임
- 실험실할때 샘플 사이즈의 결정은 
    - 보통 $ \beta $ = 0.80 와 $ \beta $ = 0.90 중에서 원하는 검정력은 선택
    - 유의수준  $ \alpha $ = 0.05 로 주어짐.
    - 표준편차
    - $ \delta $는 minimal relevant difference” (MIREDIF) 또는  “smallest meaningful difference” (SMD)라고 알려짐.

### 9.1.2 Power of two-sample t tests

$$ v = \frac{ \delta } {  \sigma  \sqrt{  1/n_1 + 1/n_2  } }   $$
